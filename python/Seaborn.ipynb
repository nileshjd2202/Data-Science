```python
df = sns.load_dataset("tips")
```

```python
df
# tip => Dependent variable
# others => Independent variable
```

```python
sns.relplot(x='total_bill',y='tip', data=df)
# graph between tip & total_bill
```

```python
sns.relplot(x='total_bill',y='tip', data=df, hue='sex')
# differentiate using sex, we use hue
```

```python
sns.relplot(x='total_bill',y='tip', data=df, hue='sex', style='time')
# differentiate using time we use style
```

```python
sns.relplot(x='total_bill',y='tip', data=df, hue='sex', style='time', size='size')
# differentiate using size
```

```python
sns.relplot(x='total_bill',y='tip', data=df, hue='sex', style='time', size='size', kind='line')
# we change graph type relplot to lineplot using kind
```

```python
sns.relplot(x='total_bill', y='tip', col='time', data=df)
```

```python
sns.catplot(x='day', y='total_bill', data=df, jitter=False)
# jitter by default true
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='swarm', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=df, hue='smoker')
```

```python
sns.catplot(x='smoker', y='tip', data=df, order=['No',"Yes"])
```

```python
sns.catplot(x='day', y='tip', kind='box', hue='sex', data=df)
```

```python
sns.catplot(x='day', y='tip', kind='boxen', data=df)
```

```python
sns.catplot(x='total_bill', y='day',data=df, kind='violin',hue='sex', split=True)
# split = combine male & female in one violin
```

```python
ax = sns.catplot(x='day',y="total_bill", kind="violin", data=df, inner=None)
sns.swarmplot(x='day', y='total_bill',data=df, color='k',size=3)
```
